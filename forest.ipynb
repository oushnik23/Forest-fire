```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=200)
```

```python
np.set_printoptions(precision=2)
```

```python
plt.scatter(y_pred,ytest)
plt.xlabel("Predicted value")
plt.ylabel("True value")
plt.title('Predicted vs True values')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print("RMSE {:.2f} :".format(np.sqrt(mean_squared_error(y_pred,ytest))))
```

```python
print("MSE {:.2f} :".format(mean_absolute_error(y_pred,ytest)))
```

```python
print("Score {:.2f} :".format(r2_score(y_pred,ytest)))
```
